```python
import json
import pandas as pd
import requests
from config import lastfm_api_key
```

```python
jsonFormat = '&format=json'
```

```python
url = 'http://ws.audioscrobbler.com/2.0/?method=chart.gettoptracks'
url_end = '&api_key=' + lastfm_api_key + jsonFormat
```

```python
limit = '5'
countries = ["Hong Kong", "Netherlands", "Russian federation", "Australia", 
             "Canada", "France", "Japan", "united kingdom", "Germany", "united states"]
song_list = []
artist_list = []
country_list = []
rank = []

for country in countries:
    top_list = []
    urlgeo = 'http://ws.audioscrobbler.com/2.0/?method=geo.gettoptracks&limit=' + limit + '&country=' + country + url_end
    try:
        lastfm_response = requests.get(urlgeo).json()
        top_list.append(lastfm_response["tracks"]["track"])
        temp_df = pd.DataFrame(top_list)
        for x in range (0,5):
            rank.append(x+1)
            song_list.append(temp_df[x][0]['name'])
            country_list.append(country)
            artist_list.append(temp_df[x][0]['artist']['name'])
    except:
        print(f"Couldn't find {country}")
```

```python
top_df=pd.DataFrame({
    "Country" : country_list,
    "Artist(s)" : artist_list,
    "Song Title" : song_list,
    "Rank" : rank
})

top_df["Source"] = "LastFM"

top_df
```

```python
albums = []
for index, row in top_df.iterrows():
    track_url = f"http://ws.audioscrobbler.com/2.0/?method=track.getInfo&artist={row['Artist(s)']}&track={row['Song Title']}&autocorrect=1" + url_end
    lastfm_response = requests.get(track_url).json()
    albums.append(lastfm_response['track']['album']['title'])
```

```python
lastfm_response['track']['album']['title']
```

```python
top_df["Album"] = albums
```

```python
top_df.replace(to_replace="Russian federation", value="Russia", inplace=True)
top_df.replace(to_replace="united kingdom", value="UK", inplace=True)
top_df.replace(to_replace="united states", value="US", inplace=True)
```
