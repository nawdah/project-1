```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
merge_df = pd.read_csv('Resources/track_analysis.csv')
merge_df
```

```python
country_df = merge_df.groupby(["Country"])
mean_value = country_df.mean()
country_mean = mean_value.reset_index()

country_mean
```

```python
x_axis = np.arange(len(country_mean))
tick_locations = [value for value in x_axis]
fig,ax=plt.subplots(figsize=(50,20))

bar_1=ax.bar(x_axis+0.10, country_mean["danceability"], label="danceability", color='r', width=0.10)
bar_2=ax.bar(x_axis+0.20, country_mean["energy"],label="energy", color='g', width=0.10)
bar_3=ax.bar(x_axis+0.30, country_mean["speechiness"], label="speechiness",color='b', width=0.10)
bar_4=ax.bar(x_axis+0.40, country_mean["acousticness"], label="acousticness", color='y', width=0.10)
bar_5=ax.bar(x_axis+0.50, country_mean["instrumentalness"], label="instrumentalness",color='purple', width=0.10)
bar_6=ax.bar(x_axis+0.60, country_mean["liveness"],label="liveness", color='c', width=0.10)
bar_7=ax.bar(x_axis+0.70, country_mean["valence"],label="valence",  color='darkorange', width=0.10)

plt.legend()
plt.grid()

def autolabel(rects):
    for p in rects:
        height = '{:.2f}%'.format(p.get_height()*100)
#         print(height)
        x = p.get_x() + p.get_width()/2-.05
        y = p.get_y() + p.get_height()+.01
#         print(x)
#         current_ax.text(p.get_x() + p.get_width()/2,.5*height,"%d" %float(height)+ "%",ha="center",va="bottom",color="white")
        ax.annotate(height, (x, y), color='k')
autolabel(bar_1)
autolabel(bar_2)
autolabel(bar_3)
autolabel(bar_4)
autolabel(bar_5)
autolabel(bar_6)
autolabel(bar_7)
```
