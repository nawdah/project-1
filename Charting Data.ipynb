```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
merge_df = pd.read_csv('Resources/track_analysis.csv')
merge_df
```

```python
country_df = merge_df.groupby(["Country"])
mean_value = country_df.mean()
country_mean = mean_value.reset_index()

country_mean
```

```python
x_axis = np.arange(len(country_mean))
tick_locations = [value for value in x_axis]
fig,ax=plt.subplots(figsize=(50,20))

bar_1=ax.bar(x_axis+0.10, country_mean["danceability"], label="danceability", color='r', width=0.10)
bar_2=ax.bar(x_axis+0.20, country_mean["energy"],label="energy", color='g', width=0.10)
bar_3=ax.bar(x_axis+0.30, country_mean["speechiness"], label="speechiness",color='b', width=0.10)
bar_4=ax.bar(x_axis+0.40, country_mean["acousticness"], label="acousticness", color='y', width=0.10)
bar_5=ax.bar(x_axis+0.50, country_mean["instrumentalness"], label="instrumentalness",color='purple', width=0.10)
bar_6=ax.bar(x_axis+0.60, country_mean["liveness"],label="liveness", color='c', width=0.10)
bar_7=ax.bar(x_axis+0.70, country_mean["valence"],label="valence",  color='darkorange', width=0.10)

plt.legend()
plt.grid()

def autolabel(rects):
    for p in rects:
        height = '{:.2f}%'.format(p.get_height()*100)
        x = p.get_x() + p.get_width()/2-.05
        y = p.get_y() + p.get_height()+.01
        ax.annotate(height, (x, y), color='k')

autolabel(bar_1)
autolabel(bar_2)
autolabel(bar_3)
autolabel(bar_4)
autolabel(bar_5)
autolabel(bar_6)
autolabel(bar_7)


plt.xticks(tick_locations, country_mean["Country"], rotation="vertical")
plt.savefig("avg_feature_country.png")
```

```python
country_id = []
for index, row in merge_df.iterrows():
    try:
        if (row["Country"] == "Australia"):
            country_id.append(0)
        elif (row["Country"] == "Canada"):
            country_id.append(1)
        elif (row["Country"] == "Denmark"):
            country_id.append(2)
        elif (row["Country"] == "France"):
            country_id.append(3)
        elif (row["Country"] == "Germany"):
            country_id.append(4)
        elif (row["Country"] == "Hong Kong"):
            country_id.append(5)
        elif (row["Country"] == "Japan"):
            country_id.append(6)
        elif (row["Country"] == "Netherlands"):
            country_id.append(7)
        elif (row["Country"] == "Russia"):
            country_id.append(8)
        elif (row["Country"] == "UK"):
            country_id.append(9)
        elif (row["Country"] == "US"):
            country_id.append(10)
        else:
            country_id.append(100)
            
    except:
        country_id.append(100)
```

```python
dance_df = merge_df
dance_df["Country_ID"] = country_id
ax = dance_df.plot(kind='scatter', x='Country_ID', y='danceability')
```

```python
artist = merge_df.groupby(['Artist']).apply(lambda x: x.count())

# Draw plot
fig, ax = plt.subplots(figsize=(12,8), dpi= 80)
ax.vlines(x=artist.index, ymin=0, ymax=artist.Artist, color='c', alpha=0.7, linewidth=2)
ax.scatter(x=artist.index, y=artist.Artist, s=75, color='c', alpha=0.7)

# Title, Label, Ticks and Ylim
ax.set_title('Frequency of Artist Occurrences', fontdict={'size':22})
ax.set_ylabel('Frequency')
ax.set_xticks(artist.index)
ax.set_xticklabels(artist.index, rotation=60, fontdict={'horizontalalignment': 'right', 'size':12})
ax.set_ylim(0, 25)

for row in artist.itertuples():
    ax.text(row.Index, row.Artist+.5, s=round(row.Artist, 2), horizontalalignment= 'center', verticalalignment='bottom', fontsize=8)

plt.show()
```

```python
songs = merge_df.groupby(['Name']).apply(lambda x: x.count())

# Draw plot
fig, ax = plt.subplots(figsize=(12,8), dpi= 80)
ax.vlines(x=songs.index, ymin=0, ymax=songs.Name, color='m', alpha=0.7, linewidth=2)
ax.scatter(x=songs.index, y=songs.Name, s=75, color='m', alpha=0.7)

# Title, Label, Ticks and Ylim
ax.set_title('Frequency of Track Occurrences', fontdict={'size':22})
ax.set_ylabel('Frequency')
ax.set_xticks(songs.index)
ax.set_xticklabels(songs.index, rotation=60, fontdict={'horizontalalignment': 'right', 'size':12})
ax.set_ylim(0, 15)

for row in songs.itertuples():
    ax.text(row.Index, row.Name+.5, s=round(row.Name, 2), horizontalalignment= 'center', verticalalignment='bottom', fontsize=8)

plt.show()
```
